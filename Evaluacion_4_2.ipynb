## Tarea 5

Get a polygons map of the lowest administrative unit possible.

Get a table of variables for those units. At least 3 numerical variables.

Preprocess both tables and get them ready for merging.

Do the merging, making the changes needed so that you keep the most columns.


```python
# En este caso, se consiguieron los mapas para los distritos, pero no se encontraron datos para cada distrito. En este sentido,
# solo quedó trabajar con las provincias. La República Democrática del Congo publicó el 2020 un anuario con muchos datos, pero
# estos casi completamente en función de las provincias. Se decidió trabajar con estos datos. 
# fuente de los datos: https://ins.cd/wp-content/uploads/2022/06/ANNUAIRE-STATISTIQUE-2020.pdf
# Estos datos estaban en pdf, así que se tuvieron que copiar manualmente a un archivo de excel
```

```python
# Se definen las provincias a partir de una shape file
import os
import pandas as pd 
import geopandas as gpd

Provincias_DRC = gpd.read_file(os.path.join("data","provincias_DRC","cod_admbnda_adm1_rgc_itos_20190911.shp"))
Provincias_DRC
```

```python
# Ahora, se leen los datos que se obtuvieron
Estadisticas_DRC = pd.read_excel(os.path.join('data', 'Estad.xlsx'))
Estadisticas_DRC
```

```python
# Hay datos irrelevantes, así que se procesan los data frame para que quede solo lo útil
```

```python
# La tabla con los datos está bastan bien, así que se deja así, pero igual se pueden cambiar los espacios de los nombres de las
# columnas
Estadisticas_DRC.columns = Estadisticas_DRC.columns.str.replace(' ', '_')
```

```python
# La gdf con los distritos sí tiene algunas columnas innecesarias

P_DRC = Provincias_DRC.loc[:,['ADM1_FR','geometry']]

#Finalmente, también se puede cambiar el nombre de la columna a provincias
P_DRC = P_DRC.rename(columns={'ADM1_FR': 'Provincia'})
P_DRC
```

```python
# Ahora sí, se puede hacer el merge
```

```python
# Se procesan los nombres
# Primero, se eliminarán los espacios
P_DRC['Provincia'] = P_DRC.Provincia.str.replace(' ','_')
Estadisticas_DRC['Provincia'] = Estadisticas_DRC.Provincia.str.replace(' ','_')
# Después, los guiones medios
P_DRC['Provincia'] = P_DRC.Provincia.str.replace('-','_')
Estadisticas_DRC['Provincia'] = Estadisticas_DRC.Provincia.str.replace('-','_')

# Después, se strippearán las columnas
Estadisticas_DRC = Estadisticas_DRC.rename(columns=lambda x: x.strip())
P_DRC = P_DRC.rename(columns=lambda x: x.strip())

# Después, se le aplica directamente el unidecode, porque los nombres en francés tienen carácteres que pueden ser molestos
import unidecode

byePunctuation=lambda x: unidecode.unidecode(x)
P_DRC[['Provincia']]=P_DRC[['Provincia']].map(byePunctuation)
Estadisticas_DRC[['Provincia']]=Estadisticas_DRC[['Provincia']].map(byePunctuation)

# Tratamos directamente con el fuzzy merge
!pip install thefuzz
from thefuzz import process as fz
# Se crean sets con los  nombres
sP = set(P_DRC.Provincia)
sD = set(Estadisticas_DRC.Provincia)
# Ahora se calculan los coeficientes de igualdad
dMerge = [(f,fz.extractOne(f, sP)) for f in sorted(sD)]
dMerge, len(dMerge)
```

```python
# Se mantuvo el número de filas?
len(dMerge) == len(P_DRC.Provincia)
```

```python
# Ahora sí se procede con el merge
DRC_Datos = Estadisticas_DRC.merge(P_DRC, how='left')
# El número de filas terminó siendo igual?
len(DRC_Datos.Provincia) == len(P_DRC.Provincia)
```

```python
# Parece que ya está mergeado de forma correcta, entonces
DRC_Datos
```

```python
# Finalmente, nos aseguramos de que el resultado sea un gdf con el crc deseado
DRC_Datos= gpd.GeoDataFrame(DRC_Datos, crs = 4326)
DRC_Datos = DRC_Datos.to_crs(32734)
type(DRC_Datos)
```

```python
# Primero, hay que probar varios sistemas para los intervalos, según el método que se enseñó en clase
# El seaborn no parece ser necesario, pero igual lo estoy instalando en caso pueda faltar algo más adelante
!pip install seaborn
import seaborn as sea
!pip install numba
import mapclassify 
import numpy as np

np.random.seed(1234)
```

```python
# Se decide hacer 6 intervalos

K=6
Variable=DRC_Datos['Esperanza_de_vida']
ei6 = mapclassify.EqualInterval(Variable, k=K)
msd = mapclassify.StdMean(Variable)                             
q6=mapclassify.Quantiles(Variable,k=K)
mb6 = mapclassify.MaximumBreaks(Variable, k=K)
ht = mapclassify.HeadTailBreaks(Variable)
fj6 = mapclassify.FisherJenks(Variable, k=K)
jc6 = mapclassify.JenksCaspall(Variable, k=K)
mp6 = mapclassify.MaxP(Variable, k=K)
```

```python
# Para comparar a las clasificaciones, se calcula la desviación absoluta media del parámetro (ADCM)
Clase = q6, ei6,msd, ht, mb6, fj6, jc6, mp6
Fits = np.array([ c.adcm for c in Clase])
ADCM = pd.DataFrame(Fits)
ADCM['Clasificador'] = [c.name for c in Clase]

ADCM.rename(columns = {0:'ADCM'},inplace=True)

ADCM
```

```python
# El ADCM del FisherJenks es el mínimo, así que se elige
```

```python
# Se crea una columna con esto valores para el gdf 
DRC_Datos['EVida_FJ'] = fj6.yb
DRC_Datos
```

```python
# Ahora, se grafica el mapa con las clasificaciones
import matplotlib.pyplot as plt
f, ax = plt.subplots(1, figsize=(9, 9))
DRC_Datos.plot(column='EVida_FJ', 
        cmap='viridis', 
        categorical=True,
        edgecolor='white', 
        linewidth=0., 
        alpha=0.75, 
        legend=True,
        legend_kwds=dict(loc=2),
        ax=ax
       )

ax.set_axis_off()
```

```python
# Finalmente, se hace el histograma
fj6_bins=list(fj6.bins)
sea.histplot(DRC_Datos.Esperanza_de_vida, bins=fj6_bins, color='blue')
```

## Tarea 7

Compute the neighbors of the capital of your country. Plot the results for each of the options.

```python
# Ahora, como no se necesitan los datos numéricos que no se pudieron conseguir para los distritos,
# sí se puede trabajar con estos
Distritos_DRC = gpd.read_file(os.path.join("data","provincias_DRC","cod_admbnda_adm2_rgc_20190911.shp"))
Distritos_DRCs= gpd.GeoDataFrame(Distritos_DRC, crs = 4326)
Distritos_DRC = Distritos_DRC.to_crs(32734)

# Después, se procesan un poco los nombres de la columna de los distritos, para no tener problemas luego
Distritos_DRC['ADM2_FR'] = Distritos_DRC.ADM2_FR.str.replace(' ','_')
Distritos_DRC = Distritos_DRC.rename(columns=lambda x: x.strip())
import unidecode
byePunctuation=lambda x: unidecode.unidecode(x)
Distritos_DRC[['ADM2_FR']]=Distritos_DRC[['ADM2_FR']].map(byePunctuation)

# Finalmente, se revisa si es un gdf
type(Distritos_DRC)
```

```python
# Se instala el libsypal y se invoca a las funciones Queen, Rook y King (las tres opciones posibles para el cálculo)
!pip install libpysal
from libpysal.weights import Queen, Rook, KNN
```

```python
# Ahora, se averigua cuál es la fila que contiene a la capital
Distritos_DRC.ADM2_FR.str.contains('Kinshasa').sort_values()
```

```python
# Es la 97!
Distritos_DRC.ADM2_FR[97]
```

```python
# Primero el rook, con los vecinos en amarillo y Kinshasa (capital) en rojo
CDR_rook = Rook.from_dataframe(Distritos_DRC,use_index=False) 
# Se grafica
base=Distritos_DRC[Distritos_DRC.ADM2_FR=="Kinshasa"].plot()
Distritos_DRC.iloc[CDR_rook.neighbors[97] ,].plot(ax=base,facecolor="yellow",edgecolor='k')
Distritos_DRC.iloc[[97],:].plot(ax=base,facecolor="red")
```

```python
# Ahora con queen
CDR_queen = Queen.from_dataframe(Distritos_DRC,use_index=False)
base=Distritos_DRC[Distritos_DRC.ADM2_FR=="Kinshasa"].plot()
Distritos_DRC.iloc[CDR_queen.neighbors[97] ,].plot(ax=base,facecolor="yellow",edgecolor='k')
Distritos_DRC.iloc[[97],:].plot(ax=base,facecolor="red")
```

```python
# Finalmente, con king
CDR_knn = KNN.from_dataframe(Distritos_DRC, k=4)
base=Distritos_DRC[Distritos_DRC.ADM2_FR=="Kinshasa"].plot()
Distritos_DRC.iloc[CDR_knn.neighbors[97] ,].plot(ax=base,facecolor="yellow",edgecolor='k')
Distritos_DRC.iloc[[97],:].plot(ax=base,facecolor="red")
```

```python
# Se observa que para rook y queen es igual, mientras que para king se pierden algunos distritos. Esto se debe a que para knn
# se fija un número de vecinos y se elige a los 'k' más cercanos. Curiosamente, en knn se tuvo a una pequeña provincia extra en vez de las
# que estaban inmediatamente cerca. 
# Parece que para este caso, queen y rook son más fiables
# Como dato final, al norte de Kinshasa está el Rio Congo y su frontera con la República del Congo (No confundir con la República Democrática del Congo)
# por lo que es normal que no haya provincias vecinas hacia el norte
```

## Tarea 8

Compute the Moran's coefficient for all your numeric variables.

Make a scatter plot for each variable.

```python
# Se calcula el índice Moran para las variables numéricas. Ahora, lamentablemente de debe de volver a trabajar con provincias
# Primero, se instalan los paquetes
!pip install esda
from esda.moran import Moran
```

```python
# Se decide usar el rook para calcular el índice Moran
# Se tienen que calcular los rooks de nuevo, porque ahora se estará trabajando con provincias
CDRP_rook = Rook.from_dataframe(DRC_Datos,use_index=False)

# Esperanza de vida
moranEV = Moran(DRC_Datos['Esperanza_de_vida'], CDRP_rook)

# Tasa de desempleo
moranTD = Moran(DRC_Datos['Tasa_de_desempleo_2016'], CDRP_rook)

# Años de escolarización
moranAE = Moran(DRC_Datos['Años_de_escolarización'], CDRP_rook)

# Promedio de años de escolarización en adultos
moranAEP = Moran(DRC_Datos['Número_promedio_de_años_de_escolarización_en_adultos'], CDRP_rook)

# Finalmente, se determinan los coeficientes
moranEV.I, moranEV.p_sim, moranTD.I, moranTD.p_sim, moranAE.I, moranAE.p_sim, moranAEP.I, moranAEP.p_sim
```

```python
# En todos los casos, el p-value es bastante alto, así que no se podría afirmar que exita clustering o no
# Sin embargo, un índice Moran positivo indicaría que hay clustering, mientras que uno negativo indicaría que hay scattering
# Probablemente, este p-value tan alto se deba a la poca cantidad de provincias. Si se hubiesen conseguido datos para los distritos,
# es posible que los resultados hayan sido mejores
```

```python
# Se plotea para Esperanza de vida
!pip install splot
from splot.esda import moran_scatterplot

fig, ax = moran_scatterplot(moranEV, aspect_equal=True)
ax.set_xlabel('EV_std')
ax.set_ylabel('SpatialLag_EV_std');
```

```python
# Tasa de desempleo
fig, ax = moran_scatterplot(moranTD, aspect_equal=True)
ax.set_xlabel('TD_std')
ax.set_ylabel('SpatialLag_TD_std');
```

```python
# Años de escolarización
fig, ax = moran_scatterplot(moranAE, aspect_equal=True)
ax.set_xlabel('AE_std')
ax.set_ylabel('SpatialLag_AE_std');
```

```python
# Años de escolarización promedio en adultos
fig, ax = moran_scatterplot(moranEV, aspect_equal=True)
ax.set_xlabel('AEP_std')
ax.set_ylabel('SpatialLag_AEP_std');
```

## Tarea 9

Compute the Local Moran for the variables in your data that have significant spatial correlation.

Create a new column for each of those variables, with a label ('0 no_sig', '1 hotSpot', '2 coldOutlier', '3 coldSpot', '4 hotOutlier').

Prepare a map for each of the variables analyzed, showing the spots and outliers.

```python
# En mi caso, ningún parámetro tuvo buena correlación espacial para las provincias. Sin embargo, se hará este trabajo para
# los datos de esperanza de vida y de años de escolarización promedio, que fueron los que presentaron los p-values más bajos
```

```python
# Se calculan los Local Moran
from esda.moran import Moran_Local

IMLEV = Moran_Local(y=DRC_Datos['Esperanza_de_vida'], w=CDRP_rook,seed=2250)
IMLAEP = Moran_Local(y=DRC_Datos['Número_promedio_de_años_de_escolarización_en_adultos'], w=CDRP_rook,seed=2250)

IMLEV.p_sim, IMLAEP.p_sim
```

```python
# Se evalúa la significancia para cada provincia, pero se aumentará un poco el margen del p-value, para compensar la
# falta de clustering en nuestros datos
DRC_Datos['EV_quadrant']=[l if p <0.08 else 0 for l,p in zip(IMLEV.q,IMLEV.p_sim)  ]

DRC_Datos['AEP_quadrant']=[l if p <0.08 else 0 for l,p in zip(IMLAEP.q,IMLAEP.p_sim)  ]

# Se cuentan
DRC_Datos['EV_quadrant'].value_counts(),DRC_Datos['AEP_quadrant'].value_counts()
```

```python
# Se crean columnas para las variables de la esperanza de vida y se cuentan los casos
labels = [ '0 no_sig', '1 hotSpot', '2 coldOutlier', '3 coldSpot', '4 hotOutlier']

DRC_Datos['EV_quadrant_names']=[labels[i] for i in DRC_Datos['EV_quadrant']]
DRC_Datos['EV_quadrant_names'].value_counts()
```

```python
# Se crean columnas para las variables de los años de escolarización promedio y se cuentan los casos
labels = [ '0 no_sig', '1 hotSpot', '2 coldOutlier', '3 coldSpot', '4 hotOutlier']

DRC_Datos['AEP_quadrant_names']=[labels[i] for i in DRC_Datos['AEP_quadrant']]
DRC_Datos['AEP_quadrant_names'].value_counts()
```
